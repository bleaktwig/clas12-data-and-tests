```python
from sympy import *
from sympy.vector import Vector
import itertools

init_printing()
```

```python
a, b, c, d, e, f, g, h, i, j, k, l, m, n, o, p = symbols("a b c d e f g h i j k l m n o p")
symlist = [a, b, c, d, e, f, g, h, i, j, k, l, m, n, o, p]
```

```python
A5 = Matrix(
    [[a,b,c,d,e],
     [b,f,g,h,i],
     [c,g,j,k,l],
     [d,h,k,m,n],
     [e,i,l,n,o]]
)

expr = A5.det()/(f*p*p + 2*b*p + a + 1)
```

```python
# (1.1)
limit(expr, p, (sqrt(-a*f + b*b - f) - b)/f)
```

```python
# (1.2)
limit(expr, p, (-sqrt(-a*f + b*b - f) - b)/f)
```

```python
# (2)
limit(limit(expr, f, 0), p, (-a-1)/(2*b))
```

```python
# (3)
limit(limit(limit(expr, a, -1), b, 0), f, 0)
```

```python
# (4)
limit(limit(expr, p, 0), a, -1)
```
