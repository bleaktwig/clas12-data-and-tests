```python
baralpha    = 0.85
plot_size = (14,8)
gray = "#c6c3ba"
contrast_colors = ["#2d3359", "#f2cf63", "#f29544", "#f26d3d", "#d93223"]
```

```python
# Read a file containing one array
def read_file(filename):
    arr = []
    f = open(filename, "r")

    for line in f:
        arr.append(float(line.replace("\n", "")))
    
    return np.array(arr)

# Scatter plots three arrays of equal size
def scatter_plot(yscale, arrs, names):
    fig, ax = plt.subplots()
    
    fig.set_size_inches(plot_size)
    index = np.arange(len(arrs[0]))

    for i in range(len(arrs)):
        plt.scatter(index, arrs[i], label=names[i])
        plt.plot(   index, arrs[i], alpha=baralpha)

    plt.yscale(yscale)
    plt.ylabel("Value")
    # plt.xticks(index, det1_00)
    plt.legend()

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(color=gray, linestyle='-', linewidth=0.25, alpha=0.5)

    plt.tight_layout()
    plt.show()
```

# Determinant Calculation Times

```python
no    = read_file("alpha_tests/no.txt")
det   = read_file("alpha_tests/det.txt")
hdet  = read_file("alpha_tests/hdet.txt")
hchol = read_file("alpha_tests/hchol.txt")
alpha = read_file("alpha_tests/alpha.txt")

# scale arrays:
for i in range(len(no)):
    det[i]   = det[i]/no[i]
    hdet[i]  = hdet[i]/no[i]
    hchol[i] = hchol[i]/no[i]
    alpha[i] = alpha[i]/no[i]

scatter_plot("linear", [det, hdet, hchol, alpha], 
             ["normal determinant", "hardcoded determinant", "hardcoded Cholesky", "alpha calculation"])
scatter_plot("linear", [det[400:450], hdet[400:450], hchol[400:450], alpha[400:450]], 
             ["normal determinant", "hardcoded determinant", "hardcoded Cholesky", "alpha calculation"])
```
