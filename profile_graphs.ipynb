## Total Computation Times:

```python
names     = ["i", "mo", "sm", "mcf", "mtf", "ma"]
names_len = np.arange(len(names))
hb_t      = np.array([189.75, 133.52, 128.49, 94.27, 78.66, 51.18])
tb_t      = np.array([ 58.18,  56.24,  48.78, 48.78, 48.78, 48.78])

# --- Plot ---
fig, ax = plt.subplots()
fig.set_size_inches(14, 8)
index = np.arange(len(names))
bar_width = 0.35
opacity = 0.5

rectsHB = plt.bar(index, hb_t, bar_width, alpha=opacity, color='b', label="DCHB")
rectsTB = plt.bar(index + bar_width, tb_t, bar_width, alpha=opacity, color='g', label="DCTB")

plt.xlabel("Engine")
plt.ylabel("Total time")
plt.title("Engines times")
plt.xticks(index + bar_width/2, names)
plt.legend()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)

plt.tight_layout()
plt.show()
```

#### Versions:
* **i**: initial
* **mo**: micro-optimizations
* **sm**: Sherman Morrison formula
* **mcf**: multithreaded cluster finding
* **mtf**: multithreaded track finding
* **ma**: both multithreaded algorithms
